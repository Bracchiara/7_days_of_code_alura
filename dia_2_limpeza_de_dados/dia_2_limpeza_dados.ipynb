```python
import pandas as pd
import numpy as np
```

```python
emprestimo_lib = pd.read_parquet('emprestimos_acervo/emprestimos_acervo.parquet')
emprestimo_lib.head()
```

```python
# Definir as condições e categorias
classificar = [(emprestimo_lib['localizacao'] < 100),
               (emprestimo_lib['localizacao'] < 200),
               (emprestimo_lib['localizacao'] < 300),
               (emprestimo_lib['localizacao'] < 400),
               (emprestimo_lib['localizacao'] < 500),
               (emprestimo_lib['localizacao'] < 600),
               (emprestimo_lib['localizacao'] < 700),
               (emprestimo_lib['localizacao'] < 800),
               (emprestimo_lib['localizacao'] < 900),
               (emprestimo_lib['localizacao'] < 1000)]

categorias = ['Generalidades.', 
              'Filosofia e psicologia', 
              'Religião', 
              'Ciências sociais', 
              'Classe vaga.', 
              'Matemática e ciências naturais',
              'Ciências aplicadas.', 
              'Belas artes', 
              'Linguagem.', 
              'Geografia. Biografia. História']

# Criar a nova coluna com np.select
emprestimo_lib['CDU'] = np.select(classificar, categorias, default = 'Não classificado')
emprestimo_lib.head()
```

```python
# Excluir a coluna registro sistema
emprestimo_lib = emprestimo_lib.drop('registro_sistema', axis=1)
emprestimo_lib.head()
```

```python
# Verificar e alterar a coluna para o formato string
emprestimo_lib['matricula_ou_siape'].dtype
```

```python
emprestimo_lib['matricula_ou_siape'] = emprestimo_lib['matricula_ou_siape'].astype('string')
emprestimo_lib['matricula_ou_siape'].dtype
```
